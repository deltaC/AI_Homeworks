```python
# Задание 4.3.2

import numpy as np
import matplotlib.pyplot as plt

X = np.arange(1, 10)
Y = (np.sqrt(1 + np.exp(np.sqrt(X)) + np.cos(np.power(X, 2))) / np.abs(1 - np.power(np.sin(X), 3))) + np.log(np.abs(2 * X))

fig, ax = plt.subplots(2)
ax[0].plot(X, Y)
ax[1].scatter(X[:5], Y[:5])
```

```python
# Задание 4.3.3

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import simps
from numpy import trapz


X = np.arange(10)
Y = np.abs(np.cos(X * np.exp(np.cos(X) + np.log(X + 1))))

fig = plt.figure()
plt.grid()
plt.plot(X, Y, c = 'black')
plt.fill_between(X, Y)

area = trapz(Y)
print(area)
```
